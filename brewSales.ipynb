```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('brews.csv')
```

```python
data.head()
```

```python
fakeNames= ['Sam', 'Melissa', 'Dave', 'Gabriela', 'Stacy']
regulars = []
def makeFakeNames(fakeArray, emptyArray):
    i = 0
    while i < 1329:
        name = random.choice(fakeArray)
        emptyArray.append(name)
        i+=1

makeFakeNames(fakeNames, regulars)
    
    
```

```python
data['name'] = regulars
```

```python
data['name'].describe()
```

```python
data.groupby(['name', 'saleType']).sum()['purchase']
```

```python
data = data.astype('object')
```

```python
uniqueBeers = data['Beer Name'].unique()[:10]
randoBeers = []

makeFakeNames(uniqueBeers, randoBeers)
```

```python
randoBeers
```

```python
uniqueBeers[:10]
```

```python
fakePrices = []

for beer in uniqueBeers:
    price = random.randint(4,13)
    fakePrices.append(price)
```

```python
fakePrices
```

```python
beerObject = dict(zip(uniqueBeers, fakePrices))
```

```python
beerObject['Arbor Cheery Cider']
```

```python
cervezas = []

makeFakeNames(uniqueBeers, cervezas)
```

```python
data['beer'] = cervezas
```

```python
def assignBeerPrices (row):
    for beer in uniqueBeers:
        if row['beer'] == beer:
            return beerObject[beer]

data['purchase'] = data.apply(lambda row: assignBeerPrices(row), axis=1)
```

```python
data = data.set_index('date')

data = data.sort_index()
```

```python
data['purchaseCount'] = data['purchase'].cumsum()
```
